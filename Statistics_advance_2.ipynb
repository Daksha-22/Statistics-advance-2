Q1. What is hypothesis testing in statistics?

Ans: Hypothesis testing in statistics is a formal procedure used to make decisions or inferences about a population based on sample data.

It helps us determine whether there is enough evidence to support a particular claim (hypothesis) about a population parameter.

Q2. : What is the null hypothesis, and how does it differ from the alternative hypothesis?

Ans: Null Hypothesis (H₀):

- The null hypothesis is the default assumption in statistics.

- It states that there is no effect, no difference, or no relationship in the population.

- It’s the claim we assume to be true unless the data provides strong evidence against it.

Example:

- A medicine has no effect on blood pressure.

- The average exam score of a class is 70.

- Two groups have the same mean.

Alternative Hypothesis (H₁ or Ha):

- The alternative hypothesis represents the claim we want to test.

- It states that there is an effect, a difference, or a relationship.

- If the evidence from the sample strongly contradicts H₀, we support H₁.


Q3.  Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Ans: The significance level, denoted by α, is the threshold we set in hypothesis testing to decide whether to reject the null hypothesis (H₀).

- It represents the probability of making a Type I error → rejecting H₀ when H₀ is actually true (a false positive).

- Common choices: 0.05 (5%), 0.01 (1%), or 0.10 (10%).

Role in Hypothesis Testing:

- Set α before testing

Example: α = 0.05 means we are willing to accept a 5% risk of wrongly rejecting H₀.

- Compare p-value with α

If p ≤ α → Reject H₀ (strong evidence against H₀).

If p > α → Fail to reject H₀ (not enough evidence against H₀).

- Decision Making

α acts as a cut-off point for determining whether results are statistically significant.

Example:

A company claims the average weight of a chocolate bar is 50g.

H₀: μ = 50

H₁: μ ≠ 50

Significance level: α = 0.05

After collecting data, you calculate the p-value = 0.03.

Since 0.03 < 0.05, you reject H₀ → There is enough evidence to suggest the average weight is not 50g.

Q4. What are Type I and Type II errors? Give examples of each.

Ans: 1. Type I Error (False Positive)

- Happens when we reject the null hypothesis (H₀) even though it is true.

- In other words, we think there is an effect/difference when in reality there isn’t.

- Probability of making this error = α (significance level).

Example:

A medical test for a new drug says the drug works (rejects H₀: “drug has no effect”), but in reality, the drug does nothing.

Reality: Drug doesn’t work (H₀ true).

Decision: We say it works (reject H₀).

Wrong conclusion = Type I error.

2. Type II Error (False Negative)

- Happens when we fail to reject the null hypothesis (H₀) even though it is false.

- In other words, we miss a real effect/difference.

- Probability of making this error = β, and (1 - β) = power of the test.

Example:

The same drug actually does work (H₁ true), but the test results say it has no effect (fail to reject H₀).

Reality: Drug works (H₀ false).

Decision: We say it doesn’t work (fail to reject H₀).

Wrong conclusion = Type II error.

Q5.  What is the difference between a Z-test and a T-test? Explain when to use each.

Ans: Z-test

Definition: A hypothesis test used when the population variance (σ²) is known and/or the sample size is large (n ≥ 30).

Assumption: Data follows a normal distribution (or approximately normal by Central Limit Theorem for large n).

Distribution used: Standard Normal Distribution (Z-distribution).

When to Use:

Population standard deviation (σ) is known.

Large sample size (n ≥ 30).

Examples:

- Testing if the mean height of students = 170 cm when σ is known.

- Checking if a coin is fair after flipping it 200 times (proportion test).

2. T-test

Definition: A hypothesis test used when the population variance (σ²) is unknown and the sample size is small (n < 30).

Assumption: Data should be approximately normally distributed.

Distribution used: Student’s t-distribution (heavier tails than normal, accounts for extra uncertainty in estimating σ).

When to Use:

- Population standard deviation (σ) is unknown.

- Small sample size (n < 30).

Examples:

- Testing if the average exam score of a class of 15 students = 70 (σ unknown).

- Comparing mean blood pressure of two small patient groups.


Q6.  Write a Python program code to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10      # number of trials
p = 0.5     # probability of success
size = 1000 # number of samples

# Generate binomial distribution data
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(0, n+2), edgecolor='black', align='left')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

Q7.  Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
mu_0 = 50

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # unbiased estimator
n = len(sample_data)

# Z-statistic
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Sample Std Dev:", sample_std)
print("Sample Size:", n)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject H0: There is significant evidence that the mean is different from 50.")
else:
    print("Fail to reject H0: No significant evidence that the mean differs from 50.")
```

Q8. Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data
np.random.seed(42)  # for reproducibility
n = 100  # sample size
mu, sigma = 50, 5  # true mean and std dev
data = np.random.normal(mu, sigma, n)

# Step 2: Calculate sample mean and std
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

# Step 3: Calculate 95% confidence interval using t-distribution
alpha = 0.05
df = n - 1
t_crit = stats.t.ppf(1 - alpha/2, df)  # two-tailed critical value
margin_error = t_crit * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_error
ci_upper = sample_mean + margin_error

print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 4: Plot the histogram
plt.hist(data, bins=15, edgecolor='black', alpha=0.7, density=True)
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='--', label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color='blue', linestyle='--', label=f"95% CI Upper = {ci_upper:.2f}")
plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

Q9. Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.


```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Function to calculate Z-scores and plot histogram.
    """
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample std dev
    z_scores = (data - mean) / std

    # Print summary
    print("Mean of data:", round(mean, 2))
    print("Standard Deviation of data:", round(std, 2))
    print("First 5 Z-scores:", np.round(z_scores[:5], 2))

    # Plot histogram of Z-scores
    plt.hist(z_scores, bins=10, edgecolor='black', alpha=0.7, density=True)
    plt.title("Histogram of Standardized Data (Z-scores)")
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.axvline(0, color='red', linestyle='--', label="Mean (Z=0)")
    plt.legend()
    plt.show()

    return z_scores

# Example usage
data = np.array([49, 50, 51, 52, 50, 48, 53, 47, 50, 49])
z_scores = calculate_z_scores(data)
```
